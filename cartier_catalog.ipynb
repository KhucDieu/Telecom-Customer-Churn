```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cartier = pd.read_csv("cartier_catalog.csv")
```

```python
cartier.head(10)
```

```python
cartier.tags.value_counts()
```

```python
cartier['str_split']= cartier['tags'].str.split(',')
```

```python
cartier.head()
```

```python
cartier['str_split'] = cartier["tags"].str.split(',')
cartier['mental'] = cartier.str_split.str.get(0).str.strip()
cartier['first_gem'] = cartier.str_split.str.get(1).str.strip()
cartier['second_gem'] = cartier.str_split.str.get(2).str.strip()
cartier['third_gem'] = cartier.str_split.str.get(3).str.strip()
cartier['fourth_gem'] = cartier.str_split.str.get(4).str.strip()
cartier.fillna(0, inplace=True)

del cartier['str_split']
```

```python
cartier.head()
```

1. Loại trang sức (gem) được sử dụng nhiều nhất

```python
cartier.drop(['ref', 'image', 'tags','title','description'], axis=1, inplace= True)
```

```python
gems= pd.concat([cartier['first_gem'], cartier['second_gem'],cartier['third_gem'], cartier['fourth_gem']], axis=0)
```

```python
gems_value = gems.value_counts()[1:].to_frame()
```

```python
gems_value.reset_index(inplace=True)
```

```python
gems_value.columns =['gem_type','count']
```

```python
gems_value
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='gem_type',y='count',data=gems_value)
plt.xlabel('Gems type')
plt.ylabel('Count')
plt.title('Count of Gems')
plt.xticks(rotation= 90)
plt.show()
```

2. Kim loại nào được sử dụng nhiều nhất

```python
cartier_category_mental = cartier.groupby('categorie')['mental'].value_counts().to_frame()
```

```python
cartier_category_mental.columns = ['count']
```

```python
cartier_category_mental.reset_index(level=[0,1],inplace=True)
```

```python
cartier_category_mental
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='categorie',y='count',hue= 'mental',data=cartier_category_mental)
plt.xlabel('Jewels type')
plt.ylabel('Count')
plt.title('Count of Jewel')
plt.legend(loc ="upper left")
plt.show()
```

3. Giá trung bình của đồ trang sức phân theo kim loại là bao nhiêu?

```python
cartier_gpl = cartier.groupby(['categorie','mental'])['price'].mean().round(2).to_frame()
```

```python
cartier_gpl = cartier_gpl.reset_index()
```

```python
cartier_gpl
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='categorie',y='price',hue= 'mental',data=cartier_gpl)
plt.xlabel('Jewel type')
plt.ylabel('Price_avg')
plt.title('Mean price of Jewel')
plt.legend(loc ="upper left")
plt.show()
```

4. Giá trung bihf của đồ trang sức phân theo kim loại là bao nhiêu?

```python
categorie_gr2= cartier.groupby('mental')['price'].mean().round(2).to_frame()
```

```python
categorie_gr2.reset_index(inplace=True)
```

```python
categorie_gr2
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='mental',y='price',data=categorie_gr2)
plt.xlabel('Mental')
plt.ylabel('Price_avg')
plt.title('Mean price of Mental')
plt.show()
```

5. Có bao nhiêu trang sức (gem) trong mỗi loại sản phẩm>

```python
cartier_gr_gem = cartier.groupby('categorie')['first_gem'].value_counts().to_frame()
```

```python
cartier_gr_gem.columns =['count']
cartier_gr_gem.reset_index(level=[0,1], inplace=True)
```

```python
cartier_gr_gem = cartier_gr_gem[cartier_gr_gem!=0]
```

```python
cartier_gr_gem
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='categorie',y='count',hue= 'first_gem',data=cartier_gr_gem, palette=sns.color_palette("cubehelix",28))
plt.xlabel('Jewel type')
plt.ylabel('Count')
plt.title('Count of Jewel')
plt.legend(ncol =4,loc ="upper left")
plt.show()
```

6. Giá trung bình của trang sức (gem) trong mỗi loại sản phẩm?

```python
cartier_gr1_gem = cartier.groupby(['categorie','first_gem'])['price'].mean().round(2).to_frame()
cartier_gr1_gem  = cartier_gr1_gem.reset_index()
```

```python
cartier_gr1_gem = cartier_gr1_gem [cartier_gr1_gem["first_gem"] != 0]
```

```python
cartier_gr1_gem
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='categorie',y='price',hue= 'first_gem',data=cartier_gr1_gem, palette=sns.color_palette("cubehelix",28))
plt.xlabel('Jewels type')
plt.ylabel('Mean price')
plt.title('Mean price of Jewel')
plt.legend(ncol =4,loc ="upper left")
plt.show()
```
